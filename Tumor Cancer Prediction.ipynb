```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn import metrics
```

```python
df=pd.DataFrame(data=pd.read_csv("Tumor Cancer Prediction_Data.csv"))
df
```

```python
data_encoded=pd.get_dummies(data=df,drop_first=True)
data_encoded.drop_duplicates(inplace=True)
data_encoded
```

```python
#returns a numpy array
x = data_encoded.drop('diagnosis_M',axis=1)
x_scaled=MinMaxScaler().fit_transform(x)
```

```python
x=x_scaled
y=data_encoded["diagnosis_M"]
```

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.2, random_state=20)
logreg = LogisticRegression()
logreg.fit(x_train,y_train)
y_pred=logreg.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
x_train1, x_test1, y_train1, y_test1= train_test_split(x, y, test_size= 0.2, random_state=20)
clf = svm.SVC(kernel='linear') # Linear Kernel
clf.fit(x_train1, y_train1)
y_pred1 = clf.predict(x_test1)
print("Accuracy:",metrics.accuracy_score(y_test1, y_pred1))
```

```python
x_train2, x_test2, y_train2, y_test2= train_test_split(x, y, test_size= 0.2, random_state=20)
tree=DecisionTreeClassifier()
tree.fit(x_train2,y_train2)
y_pred2=tree.predict(x_test2)
print("Accuracy:",metrics.accuracy_score(y_test2, y_pred2))
```
